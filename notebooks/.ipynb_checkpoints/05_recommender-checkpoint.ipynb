# RECOMMENDATION ENGINE

## Product Recommendation System using Collaborative Filtering

```python
import pandas as pd
import numpy as np

from sklearn.metrics.pairwise import cosine_similarity

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

df = pd.read_excel(
    "../data/online_retail_II.xlsx"
)

df["InvoiceDate"] = pd.to_datetime(
    df["InvoiceDate"]
)

df.head()
```

```python
df.shape
df.columns
```

```python
print("Rows :", df.shape[0])
print("Columns :", df.shape[1])
```

```python
df = df[
    ~df["Description"].str.contains(
        "test",
        case=False,
        na=False
    )
]
```

### User-Item Matrix

```python
user_item_matrix = df.pivot_table(
    index="Customer ID",
    columns="Description",
    values="Quantity",
    aggfunc="sum",
    fill_value=0
)

user_item_matrix.head()
```

```python
print("User-Item Matrix Shape:")
print(user_item_matrix.shape)
```

## Customer Similarity using Cosine Similarity

```python
customer_similarity = cosine_similarity(
    user_item_matrix
)

customer_similarity_df = pd.DataFrame(
    customer_similarity,
    index=user_item_matrix.index,
    columns=user_item_matrix.index
)

customer_similarity_df.head()
```

```python
# Heatmap
plt.figure(figsize=(10,8))

sample_customers = customer_similarity_df.iloc[:20,:20]

sns.heatmap(
    sample_customers,
    cmap="YlGnBu"
)

plt.title(
    "Customer Similarity Heatmap",
    fontsize=14,
    fontweight="bold"
)

plt.tight_layout()
plt.show()
```

## Collaborative Filtering

```python
# Recommendation Function
def recommend_products(customer_id, n_recommendations=5):

    # Similar customers
    similar_customers = (
        customer_similarity_df.loc[customer_id]
        .sort_values(ascending=False)
        .iloc[1:11]
    )

    similar_customer_ids = similar_customers.index

    # Products purchased by similar customers
    recommendation_scores = (
        user_item_matrix.loc[similar_customer_ids]
        .sum(axis=0)
        .sort_values(ascending=False)
    )

    # Products already purchased by target customer
    purchased_products = set(
        user_item_matrix.loc[customer_id]
        [user_item_matrix.loc[customer_id] > 0]
        .index
    )

    # Remove already purchased products
    recommendation_scores = recommendation_scores[
        ~recommendation_scores.index.isin(purchased_products)
    ]

    # Keep only products with score > 0
    recommendation_scores = recommendation_scores[
        recommendation_scores > 0
    ]

    return recommendation_scores.head(n_recommendations)
```

```python
# Find customer with maximum purchases

top_customer = (
    user_item_matrix.sum(axis=1)
    .sort_values(ascending=False)
    .index[0]
)

print("Top Customer ID:", top_customer)

purchased = user_item_matrix.loc[top_customer]
purchased = purchased[purchased > 0]

print("Products Purchased:", len(purchased))
print("\nTop Purchased Products:")
print(purchased.head(10))
```

```python
recommendations = recommend_products(
    top_customer,
    n_recommendations=10
)
```

## Top 10 Product Recommendations

```python
import matplotlib.pyplot as plt
import seaborn as sns

recommendations = recommendations.sort_values()

plt.figure(figsize=(12,6))

ax = sns.barplot(
    x=recommendations.values,
    y=recommendations.index,
    palette="mako",   
    hue=recommendations.index,
    legend=False
)

for i, v in enumerate(recommendations.values):
    ax.text(
        v + 0.2,
        i,
        str(v),
        va="center",
        fontsize=10,
        fontweight="bold"
    )

plt.title(
    "Top 10 Recommended Products for Similar Customers",
    fontsize=18,
    fontweight="bold",
    pad=15

)

plt.xlabel(
    "Recommendation Score",
    fontsize=12
)

plt.ylabel(
    "Product",
    fontsize=12
)

plt.grid(
    axis="x",
    linestyle="--",
    alpha=0.3
)

plt.tight_layout()
plt.show()
```

## Findings

- A collaborative filtering recommendation engine was developed using customer purchase behavior.
- User-Item Matrix was created using Customer ID and Product Description.
- Customer similarity was calculated using Cosine Similarity.
- Products purchased by similar customers were used to generate recommendations.
- Top product recommendations were ranked using recommendation scores.
- The recommendation engine can be integrated into the Streamlit dashboard to provide personalized product suggestions.
- This approach helps improve customer engagement, cross-selling opportunities and customer retention.
- Top 10 recommendations were generated for a high-activity customer using collaborative filtering.
